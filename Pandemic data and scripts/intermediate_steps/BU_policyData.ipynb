```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
import os
os.chdir("../raw data")
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]
fips
```

```python
bu = pd.read_csv("bu.csv")
bu = bu.iloc[0:52, :]
bu.tail()
```

```python
def dateconvert(date_str):
    
    if (date_str) != '0' and date_str != '1' and date_str != 0 and date_str != 1 :
        diff = pd.Timestamp(date_str)
        return diff
    else:
        return (date_str)  
```

```python
emergency_declare = []
k12_close = []   
dayCare_close = []
nursing_visitorBan = []
stayAtHome_declare = []
nonEssentaialBusiness_closed = []

relegiousGathering = []
AlcoholStores_open = []
firearms_open = []
closed_restaurants = []
closed_gyms = []
closed_movieTheatres = []
Froze_evictions = []
Froze_utilityShutoff = []
Froze_mortgagePayments = []
Medical_expansion = []
Paid_sick_leaves = []


for i in fips.iloc[:, 0]:
    emergency_declare.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 2]))
    k12_close.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 3]))
    dayCare_close.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 4]))    
    nursing_visitorBan.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 5]))
    stayAtHome_declare.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 6]))
    nonEssentaialBusiness_closed.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 7]))    
    relegiousGathering.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 8]))
    AlcoholStores_open.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 9]))
    firearms_open.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 10]))
    closed_restaurants.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 11]))
    closed_gyms.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 12]))
    closed_movieTheatres.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 13]))
    Froze_evictions.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 14]))
    Froze_utilityShutoff.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 15]))
    Froze_mortgagePayments.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 16]))
    Medical_expansion.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 17]))
    Paid_sick_leaves.append(dateconvert(bu.loc[bu['State_FIPS'] == i].iloc[0, 18]))
```

```python
fips["emergency"] = emergency_declare
fips["k12"] = k12_close
fips["dayCare"] = dayCare_close
fips["nursing_visitorBan"] = nursing_visitorBan
fips["stayAtHome_declare"] = stayAtHome_declare
fips["nonEssentaialBusiness_closed"] = nonEssentaialBusiness_closed
fips["closed_restaurants"] = closed_restaurants
fips["closed_gyms"] = closed_gyms
fips["closed_movieTheatres"] = closed_movieTheatres
fips["Froze_evictions"] = Froze_evictions
fips["Froze_utilityShutoff"] = Froze_utilityShutoff
fips["Froze_mortgagePayments"] = Froze_mortgagePayments

fips["relegiousGathering"] = relegiousGathering
fips["AlcoholStores_open"] = AlcoholStores_open
fips["firearms_open"] = firearms_open
fips["Medical_expansion"] = Medical_expansion
fips["Paid_sick_leaves"] = Paid_sick_leaves
fips
```
