```python
JHU_caseAndDeath = pd.read_csv("JHU_caseAndDeath.csv")
JHU_caseAndDeath = JHU_caseAndDeath.iloc[:, 7:]
JHU_caseAndDeath.head()


BU_policyData = pd.read_csv("BU_policyData.csv")
BU_policyData = BU_policyData.drop(BU_policyData[BU_policyData.fips == '0'].index)
BU_policyData = BU_policyData.drop(BU_policyData[BU_policyData.fips == 'unknown'].index)
BU_policyData = BU_policyData.reset_index(drop=True)
BU_policyData = BU_policyData.iloc[:, 6:]
BU_policyData.head()

mobility_descarteslab = pd.read_csv("mobility_descareteslab.csv")
mobility_descarteslab = mobility_descarteslab.iloc[:, 6:]
mobility_descarteslab.head()


mobility_google = pd.read_csv("mobility_google.csv")
mobility_google = mobility_google.iloc[:, 6:]


mobility_unacast = pd.read_csv("mobility_unacast.csv")
mobility_unacast = mobility_unacast.iloc[:, 6:]
mobility_unacast.head()


covidtrackingproject = pd.read_csv("covidtrackingproject.csv")
covidtrackingproject = covidtrackingproject.iloc[:, 6:]
covidtrackingproject.head()


govex_hospital_data = pd.read_csv("govex_hospital_data.csv")
govex_hospital_data = govex_hospital_data.iloc[:, 6:]
govex_hospital_data.head()

govex_restrictions_data = pd.read_csv("govex_restrictions_data.csv")
govex_restrictions_data = govex_restrictions_data.iloc[:, 6:]
govex_restrictions_data.head()



```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])

fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == 'unknown'].index)
fips = fips.reset_index(drop=True)
```

```python
fips2 = pd.concat([fips, mobility_descarteslab, mobility_unacast, mobility_google], axis=1)
fips2 = fips2.drop(fips2[fips2.fips == 0].index)
fips2 = fips2.drop(fips2[fips2.fips == 'unknown'].index)
fips2 = fips2.reset_index(drop=True)
```

```python
new = []
for col in fips2.columns:
    arr = col.split("_")
    n = len(arr)
    if n == 1:
        new.append(col)
    else:
        date = arr[0]
        name = ""
        for header in arr[1:]:
            if name == "":
                name = header
            else:
                name = name + "_" + header
        new.append(name + "_" + date)
fips2.columns = new
fips2 = fips2.iloc[0:3142]
fips2.to_csv("Mobility Data.csv")
fips2
```
