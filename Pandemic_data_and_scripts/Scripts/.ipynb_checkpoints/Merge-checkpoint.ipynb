```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
```

```python
import os
os.chdir("../Raw data")
```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])
```

```python
os.chdir("../Intermediate data")
```

```python

JHU_caseAndDeath = pd.read_csv("JHU_caseAndDeath.csv")
JHU_caseAndDeath = JHU_caseAndDeath.iloc[:, 7:]
JHU_caseAndDeath.head()

BU_policyData = pd.read_csv("BU_policyData.csv")
BU_policyData = BU_policyData.iloc[:, 7:]
BU_policyData.head()
JHU_caseAndDeath

```

```python
fips = pd.concat([fips, BU_policyData, JHU_caseAndDeath], axis=1)
fips
```

```python
fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == 'unknown'].index)
fips
```

```python
fips.to_csv("Pandemic_v1.csv")
```
