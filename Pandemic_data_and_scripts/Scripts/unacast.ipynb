```python
import pandas as pd
from datetime import datetime
```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])

fips = fips.drop(fips[fips.fips == 0].index)
fips = fips.drop(fips[fips.fips == "unknown"].index)
fips.iloc[3142]
fips = fips.reset_index(drop=True)
fips = fips[0:3142]
fips.head()
```

```python
df = pd.read_csv("sds-v3-full-county.csv")
```

```python
df.head()
```

```python
for i in range(154505, df.shape[0]):
    old = df.iloc[i, 7]
    dto = datetime.strptime(old, '%Y-%m-%d')
    df.iloc[i, 7] = dto.strftime('%Y%m%d')
```

```python
dates = []
for date in df.iloc[:, 7]:
    if date not in dates:
        dates.append(date)
dates.sort()
#dates
```

```python
cols = df.columns[5:]
cols
```

```python
for date in dates:
    for col in cols:
        fips[date + "_" + col] = ""
```

```python
for i in range(df.shape[0]):
    for col in df.columns[5:]:
        fips.loc[(fips["fips"] == df.iloc[i, 3]), (df.iloc[i, 7] + "_" + col)] = df.loc[(i), (col)]
        
    
```
