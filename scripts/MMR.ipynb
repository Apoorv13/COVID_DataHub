```python
import pandas as pd
import numpy as np
from datetime import datetime
import time
```

```python
import os
os.chdir("../raw data")
```

```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]
fips.head()

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])
```

```python
# importing the 4 mmr data
cardiovascular = pd.read_excel('cardiovascular.xlsx')
cardiovascular = cardiovascular.iloc[:, 2:]
cardiovascular.head()

diabetes = pd.read_excel('diabetes.xlsx')
diabetes = diabetes.iloc[:, 2:]
diabetes.head()

mental = pd.read_excel('mental.xlsx')
mental = mental.iloc[:, 2:]
mental.head()

neoplasm = pd.read_excel('neoplasm.xlsx')
neoplasm = neoplasm.iloc[:, 2:]
neoplasm.head()
```

```python
def row_loc_county(inp):
    location = fips.index[(fips['fips']) == inp].values
    location = location.tolist()
    return location
```

```python
fips["c_population"] = ""
fips["cardiovascular_deaths"] = ""
fips["cardiovascular_crudeRate"] = ""

fips["d_population"] = ""
fips["diabetes_deaths"] = ""
fips["diabetes_crudeRate"] = ""

fips["m_population"] = ""
fips["mental_deaths"] = ""
fips["mental_crudeRate"] = ""

fips["n_population"] = ""
fips["neoplasm_deaths"] = ""
fips["neoplasm_crudeRate"] = ""

fips.head()
```

```python
for i in range(cardiovascular.shape[0]):
    fips.iloc[row_loc_county(cardiovascular.iloc[i, 2]), fips.columns.get_loc("cardiovascular_deaths")] = cardiovascular.iloc[i, 3]
    fips.iloc[row_loc_county(cardiovascular.iloc[i, 2]), fips.columns.get_loc("c_population")] = cardiovascular.iloc[i, 4]
    fips.iloc[row_loc_county(cardiovascular.iloc[i, 2]), fips.columns.get_loc("cardiovascular_crudeRate")] = cardiovascular.iloc[i, 5]

for i in range(diabetes.shape[0]):
    fips.iloc[row_loc_county(diabetes.iloc[i, 2]), fips.columns.get_loc("diabetes_deaths")] = diabetes.iloc[i, 3]
    fips.iloc[row_loc_county(diabetes.iloc[i, 2]), fips.columns.get_loc("d_population")] = diabetes.iloc[i, 4]
    fips.iloc[row_loc_county(diabetes.iloc[i, 2]), fips.columns.get_loc("diabetes_crudeRate")] = diabetes.iloc[i, 5]

for i in range(mental.shape[0]):
    fips.iloc[row_loc_county(mental.iloc[i, 2]), fips.columns.get_loc("mental_deaths")] = mental.iloc[i, 3]
    fips.iloc[row_loc_county(mental.iloc[i, 2]), fips.columns.get_loc("m_population")] = mental.iloc[i, 4]
    fips.iloc[row_loc_county(mental.iloc[i, 2]), fips.columns.get_loc("mental_crudeRate")] = mental.iloc[i, 5]
    
for i in range(neoplasm.shape[0]):
    fips.iloc[row_loc_county(neoplasm.iloc[i, 2]), fips.columns.get_loc("neoplasm_deaths")] = neoplasm.iloc[i, 3]
    fips.iloc[row_loc_county(neoplasm.iloc[i, 2]), fips.columns.get_loc("n_population")] = neoplasm.iloc[i, 4]
    fips.iloc[row_loc_county(neoplasm.iloc[i, 2]), fips.columns.get_loc("neoplasm_crudeRate")] = neoplasm.iloc[i, 5]
    


```
